## NASA turbofan dataset

### Introducción:

Publicado en el año 2008, este dataset se origina de un concurso lanzado en la Primera "International Conference on Prognostics and Health Management (PHM08)", y cuya data fue generada mediante C-MAPSS, una herramienta de MATLAB/Simulink desarrollada por la NASA para simular motores turbofan comerciales.

Al estar citado este dataset por más de 1000 investigaciones, se ha convertido en el "hello world" para probar los algoritmos de mantenimiento predictivo, siendo un campo de gran interés de los presentes autores, especializados como ingenieros técnicos industriales.

### Definición del problema:

El turbofan dataset se compone de 4 datasets de menor a mayor complejidad, siendo el primer dataset FD001 el punto de partida para practicar análisis predictivo, ya que establece una sola condición operacional, contempla un sólo tipo de fallo, y tiene menos ruido en la data.

La simulación en el dataset FD001 establece 100 motores turbofan que operan normalmente bajo condiciones similares. Para el dataset de entrenamiento, cada uno de los motores funciona en unidades de tiempo llamados ciclos, hasta que cada uno desarrolle una falla. Para el dataset de prueba, se registra la data hasta un tiempo indeterminado, siendo la meta predecir cuántos ciclos quedan para cada motor hasta que falle.
Las variables suministradas en el dataset FD001 son las siguientes:
- Número de motor
- Tiempo, en ciclos
- Settings operacionales
- Lecturas de 21 sensores.

Como se ha establecido, la meta es predecir la vida útil remanente (RUL - Remaining Useful Life) para cada uno de los motores.

### Marco teórico:

El mantenimiento predictivo constituye una estrategia de gestión de activos que busca anticipar el fallo de un componente o sistema antes de que ocurra, optimizando así los intervalos de intervención y reduciendo tanto el tiempo de inactividad no planificado como los costes operativos.

En este contexto surge el campo de la Gestión del Pronóstico y la Salud (Prognostics and Health Management, PHM), cuyo objetivo es integrar la adquisición de datos de sensores, el procesamiento de señales, la detección de anomalías y la estimación de la vida útil remanente en un flujo de trabajo unificado orientado a la toma de decisiones de mantenimiento.

Vida Útil Remanente (RUL)

La Vida Útil Remanente (Remaining Useful Life, RUL) es la variable objetivo central en los problemas de pronóstico de degradación. Se define formalmente como el número de ciclos operativos —o unidades de tiempo— que le restan a un componente antes de alcanzar un umbral de fallo predefinido, dado su estado de degradación actual.

Modelos de Datos para el Pronóstico de Degradación:

Los enfoques de modelado en PHM se clasifican en tres grandes categorías:

- Modelos basados en física
- Modelos basados en datos
- Modelos híbridos

El presente análisis se enmarca en el enfoque basado en datos, cuya cadena de procesamiento típica comprende: adquisición de señales de sensores → preprocesamiento y selección de características → entrenamiento del modelo de regresión → evaluación mediante métricas de error.
